```
import pandas as pd

df_csv = pd.read_csv(
    "employees.csv",
    sep=",",
    na_values=["NULL", "?"],
)
print("csv:",df_csv.head())

print(df_csv.set_index("employee_id"))
print(df_csv.reset_index())

high_salary = df_csv[df_csv["salary"] > 60000]
print(high_salary)


df_csv["salary"] = df_csv["salary"] * 1.1    # 10% raise
df_csv.to_csv("employees_updated.csv", index=False)

files.download("employees_updated.csv")
```

```
import pandas as pd

df_excel = pd.read_excel(
    "students.xlsx",
    sheet_name="students_info",
)

print("excel: ",df_excel.head())

df_json = pd.read_json(
    "products.json",
    orient="records",
)

print("json:",df_json.head())

df_excel["age"] = df_excel["age"] + 1
df_json["price"] = df_json["price"] * 1.2

# Save Excel
df_excel.to_excel("students_updated.xlsx", index=False)

# Save JSON
df_json.to_json("products_updated.json", orient="records", indent=4)

# Download both files
from google.colab import files
files.download("students_updated.xlsx")
files.download("products_updated.json")
```
